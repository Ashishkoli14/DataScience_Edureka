```python
import numpy as np
n_samples = 5000
X = np.concatenate((
    np.random.normal((-2, -2), size=(n_samples, 2)),
    np.random.normal((2, 2), size=(n_samples, 2))
))
```

```python
X
```

```python
fcm = FCM(n_clusters=2)
fcm.fit(X)
```

```python
fcm.centers
```

```python
fcm.predict(X)
```

```python
fcm_clusters = fcm.centers
fcm_labels = fcm.predict(X)
f,axes = plt.subplots(1,2,figsize=(11,5))
axes[0].scatter(X[:,0],X[:,1],alpha=0.1)
axes[1].scatter(X[:,0],X[:,1],c = fcm_labels,alpha=.1)
axes[1].scatter(fcm_clusters[:,0],fcm_clusters[:,1],marker='+',s=500,c='w')
plt.show()
```
