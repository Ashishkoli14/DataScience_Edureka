**Q2. Problem Statement: Data Filtering using Pandas** <br>
Use the above data set and do the below tasks.
1. Remove the data of players who didn't  score a single century.
2. Replace the player name "V Kohli " with "Virat Kohli '.
3. How many matches they weren't getting a chance for batting to fetch that and stored in a new column and named it as “not batting”.
4. Also fetch data where player scored neither 100 nor 50 in matches and store in the new column as renamed as "low score innings".



**Step-1:** Droping the column contain 0 as data in '100s' column


```python
#1. remove the data of plyers who didn't even score single century.
df.drop(df[df['100s'] == 0].index, inplace = True)
df
```

**Step-2:** Replace name using replace function

```python
#2 replace the player name "V Kohli " with "Virat Kohli '.
df["name"] = df["name"].replace(to_replace = "V Kohli",value ="Virat Kohli")
df
```

**Step-3:** Create new column and stored data as per given reqirement

```python
#3 how many matches they weren't getting a chance for batting, fetch that and stored in a new column and named it as “not batting”.
df["not batting "] = df["matches"] - df["innings"]
df
```

**Step-4:** Create new column and stored data as per given reqirement

```python
#4 also fetch data where they scored neither 100 nor 50 in matches and store in the new column as renamed as "low score innings".
df["low score innings"] = df["innings"]-df["100s"]-df["50s"]
df
```
