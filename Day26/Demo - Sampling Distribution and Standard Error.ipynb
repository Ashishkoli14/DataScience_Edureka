## Population and Sample
<br>
<strong>Population: </strong>Entire set of data on which we want to draw conclusions<br>
<strong>Sample: </strong>Collection of observations drawn from a population. It is a subset of population
<br><br>

### Consider the given dataset of students as population, extract the age column and find its parameters (mean, variance and standard deviation)

The above data is the data of the university, let us assume it as a whole population<br>
Let's calculate the population parameters of age data from the above data.

```python
# Finding the population parameters of age

# Extracting 'age' data from the above data
age = data.age

# population mean
mu = age.mean()
print('Population Mean of Age:',mu)

# population standard deviation
sigma = age.std()
print('Population Standard deviation of Age:',sigma)

# population variance
sigma_square = age.var()
print('Population Variance of Age:',sigma_square)
```

We have calculated the population parameters. Sometimes it is hard to find the population because
    - It may be expensive to collect data or
    - It may be impossible to collect the data in the short time

### Take a sample of age with sample size (n) = 1000 from the population and find its statistics (mean, variance and standard deviation) and state whether the statistics calculated from the sample are good estimations of population parameters or not. 

Now let's have a look at how to take samples from the population and calculate the statistics

```python
# Taking a sample of 1000 students from the data provided

sample_data = data.sample(1000, random_state = 1)
print('Number of instances in the sample:',sample_data.shape[0])
```

```python
# Calculating the statistics of the sample

# Extracting age data from the sample data
sample_data_age = sample_data.age

# Calculating the mean of the sample
x_bar = sample_data_age.mean()
print('Mean of age from sample data:',x_bar)

# Calculating the standard deviation of sample
s = sample_data_age.std()
print('Standard deviation of age from sample data:',s)

# Calculating the variance of sample
s_square = sample_data_age.var()
print('Variance of age from sample data:',s_square)
```

### Inference
The statistics calculated from the sample are good estimates of population parameters because there is minute difference in the values of statistics and parameters.

## Sampling Distribution
Sampling distribution is the distribution of a statistic obtained from several samples,
    - Consider different samples of data, let 'n' be the number of samples
    - Calculate the statistic of every sample
    - The distribution of values of the statistic obtained from 'n' samples is known as Sampling Distribution

### Consider 20 samples of 'nc.score' with sample size of 200, find the sampling distribution of mean 

```python
# Let us extract the nc.score column from the data
score = data['nc.score']
```

```python
# Calculating means of each sample
means = list()

for i in range(20):
    sample_score = score.sample(200)
    mean = sample_score.mean()
    means.append(mean)
    
print('Means of all the samples:',means)
```

```python
# Let's take a look at the sampling distribution of mean

sns.distplot(means)
```

### Inference
The above distribution is the <b>Sampling Distribution of Mean</b> with 20 samples of sample size 200

## Standard Error
Standard Error of a statistic is defined as the variation of the sample statitic to its population parameter. It helps to find how much a sample represents the population.

### Standard Error with one sample

### Consider the weights of students in university, take a sample, calculate the standard error of the mean of that sample and describe variation of sample mean with the population mean

```python
# Taking a sample of weights data from the data with sample size n = 500
sample_weight = data.weight.sample(500)

# Mean of the sample
sample_weight_mean = sample_weight.mean()
print('The Mean of the sample is:',sample_weight_mean)
```

We know that,

## $Standard\ Error = \frac{s}{\sqrt{n}}$

$s$ is the standard deviation <br>
$n$ is the number of samples

```python
# Finding the standard deviation of the sample 
s = sample_weight.std()
print('Standard Deviation is:',s)

# Finding thr sample size
n = sample_weight.shape[0]
print('Number of samples is:',n)
```

```python
# calculating the Standard Error

standard_error = s/np.sqrt(n)
print('Standard Error is:',standard_error)
```

### Inference
The standard error is <b>0.39</b>. 0.39 is the standard deviation of the sample mean from the population mean.

### Standard Error with Multiple Samples

To find the Standard Error of a statistic using multiple samples we have to,
    - Collect multiple samples of data
    - Take the Sampling Distribution of the statistic
    - Find the standard deviation of the Sampling Distribution obtained
    - The Standard Deviation of the Sampling Distribution represents the Standard Error of the particular statistic
    
### Consider the weights of students in university, take 20 samples of size 200, calculate the standard error of mean of that sample and describe variation of mean of sampling distribution with the population mean
Let us take 20 samples of sample size 200, calculate the Sampling Distribution of Mean and find its standard error.

```python
means = list()

# Taking samples of weight data and finding its mean
for i in range(20):
    sample = data.weight.sample(200)
    mean = sample.mean()
    means.append(mean)
    
print('The Means of the samples:',means)
```

```python
# Sampling Distribution
sns.distplot(means)
```

```python
# Lets find the Standard Deviation of the Sampling Distribution which is also the Standard Error of the statistic

standard_error = np.array(means).std()
print('The Standard Error is:',standard_error)
```

### Inference
The standard error is 0.67. The mean of the sampling distribution is 0.67 standard deviations away from the population mean
