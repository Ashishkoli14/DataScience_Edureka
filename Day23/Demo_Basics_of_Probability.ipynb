<h2>Demo - Basics of Probaility</h2>

Probability is nothing but a measure of chance of getting a desired outcome. For example, probability of raining or probability of finding a desired room through online booking.

Probability allows to put a quantifiable number to an unquantifiable entity - chance, using mathematical rules and available data.

Let us understand basic terminology and rules of calculating Probability in this exercise.

### Import Required Libraries

```python
import pandas as pd
import numpy as np
```

### Let us calculate probabilities of various Events based on a Deck of Cards.

**_Sample Space:_** A set of all possible outcomes.
    In this case, the sample space consists of 52 cards in the deck.
    
**_Event:_** Any experiment representing a subset of sample space.

#### Q. Calculate the probability of randomly drawing an Ace.

#### Probability of an Event A = P(A) = (number of favorable outcomes)/(Size of Sample Space)

Here, the specified event is drawing an ace.

```python
aces = 4
ss = 52

p_ace = round(aces/ss,3)

print("Probability of randomly drawing an Ace = ", p_ace)
```

#### Q. Calculate the probability of randomly drawing a card of Spade.

```python
spades = 13

p_spade = round(spades/ss,3)

print("Probability of randomly drawing a card of Spade = ", p_spade)
```

#### Q. What is the percentage of getting a face card?

```python
#King, Queen and Jack are the face cards in each of the 4 suits
faces = 12

p_face = round(faces/ss, 3)
print("Probability of randomly drawing a face card = ", p_face)
print("Percentage of randomly drawing a face card = ", p_face * 100, "%")
```

### Inference:

By simple comparing the three probability values, 0.077, 0.25, 0.231, the events of drawing a card of Spade and drawing a face card are equally likely to happen. Whereas drawing an Ace is a relatively rare event.
