**Q1. Problem Statement: Handling Null Values**

Load the "titanic.csv" file provided on LMS into a DataFrame, find the coumns with null values, and drop all the rows in DataFrame with null values to create a new DataFrame.


**Step-1:**Loading the dataset in DataFrame.

```python
# importing pandas package
import pandas as pd
import numpy as np
	
# making data frame from csv file
df = pd.read_csv("/content/titanic.csv")
display(df.head())
```

**Step-2:** Identifying null values in DataFrame.

```python
print("Identifying null values:")
display(df.isnull().sum())
```

**Q2. Problem Statement: Handling Duplicate Values**

Load the "titanic.csv" file provided on LMS into a DataFrame, find out the fields in the dataset with duplicate values based on the “Ticket” column, and drop all the rows with duplicate values to create a new DataFrame.


**Step-1:**Loading the dataset in a DataFrame.

```python
# importing pandas package
import pandas as pd
import numpy as np
	
# making data frame from csv file
df = pd.read_csv("/content/titanic.csv")
display(df)
display(df.isnull())
```

**Step-2:** Dropping duplicate rows based on "Ticket" column.

```python
#Dropping the rows with duplicate values
df2 = df.drop_duplicates(subset ="Ticket")
display(df2)
```
