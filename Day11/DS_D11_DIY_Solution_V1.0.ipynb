**Q1. Problem Statement: Working with DataFrames - I**

Write a Python program that reads the "iris.csv" dataset (Available on LMS) into a DataFrame, filters the rows on the condition that “sepal length” should be more than or equal to 6cm. Write the filtered rows in a new DataFrame, and save the new DataFrame as a CSV file.


**Step-1:** Loading a CSV dataset in a DataFrame and filtering the data based on condition -"sepal length" >=6 cm. 

```python
import pandas as pd
 
# reading csv file
df1 = pd.read_csv("Iris.csv")
print("Original DataFrame is:")
display(df1)

# loc --> get value - label
# iloc --> get index position

df2 = df1.loc[df1['SepalLengthCm'] >= 6]
print("Flowers with morethan 6 cm sepal length are:")
display(df2)
```

```python
cond = df1['Species'] == "Iris-virginica"
```

```python
df1[cond]
```

**Step-2:** Saving the DataFrame as a CSV file in local file system.

```python
df1[cond].to_csv("Iris-virginica.csv")
```
