**Q1. Problem Statement: Univariate EDA**

Load the “cereal.csv” data into a DataFrame and perform the following tasks:
1.	Explore the DataFrame using info() and describe() functions
2.	Replace the manufacturer names with the dictionary of names given here –
{'A':'American Home Food Products', 'G':'General Mills', 'K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}
3.	Find out the top five cereal manufacturers based on customer ratings
4.	With the help of a barplot, compare the manufacturers based on customer ratings
5.	With the help of a pie chart, find out which of the manufacturers are the largest producers of cereals



```python
import pandas as pd

# Load the “cereal.csv” data into a DataFrame
df = pd.read_csv("cereal.csv")
display(df.head())
```

```python

# Explore the DataFrame using info() and describe() functions
df.info()
df.describe()
```

```python
# Replace the manufacturer names with the dictionary of names given here
manufacturer_dict = {'A':'American Home Food Products', 'G':'General Mills', 'K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}

# Replace the manufacturer names in the DataFrame
df['mfr'] = df['mfr'].replace(manufacturer_dict)
```

```python
# Find out the top five cereal manufacturers based on customer ratings
top_five_manufacturers = df.groupby("mfr")["rating"].mean().nlargest(5)

# Print the top five manufacturers
print(top_five_manufacturers)
```

```python
# With the help of a barplot, compare the manufacturers based on customer ratings
import matplotlib.pyplot as plt

# Create a barplot of the top five manufacturers
plt.bar(top_five_manufacturers.index, top_five_manufacturers.values)

# Label the x-axis and y-axis
plt.xlabel("Manufacturer")
plt.ylabel("Customer Rating")
plt.title('Manufacturers with Highest Customer Ratings', fontsize = 20)

# Show the plot
plt.show()
```

```python
# With the help of a pie chart, find out which of the manufacturers are the largest producers of cereals
import matplotlib.pyplot as plt

# Create a pie chart of the number of cereals produced by each manufacturer
plt.pie(df["mfr"].value_counts(), labels=df["mfr"].value_counts().index)

# Label the pie chart
plt.title("Number of Cereals Produced by Manufacturer")

# Show the pie chart
plt.show()
```

**Observations:**
Kelloggs is the largest producer of cereals.
